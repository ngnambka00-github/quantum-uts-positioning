### Visualize DATA

```python
bar_data = uts_filter_train_data["Floor_ID"].value_counts()
```

```python
display(bar_data)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.bar(bar_data.index, bar_data)
ax.set_xticks(bar_data.index)
plt.show();
```

```python
def get_floor_dict(uts_data):
    floor_pos = {}
    for i in np.unique(uts_data["Floor_ID"]):
        floor_pos[i] = uts_data[uts_data["Floor_ID"] == i]

    return floor_pos

floor_pos = get_floor_dict(uts_filter_train_data)
```

```python
display(floor_pos)
```

```python
# Visualize data in floors
fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))

for i in range(4): 
    for j in range(4):
        floor_idx = floor_pos[i * 4 + j - 3]
        ax[i, j].scatter(floor_idx["Pos_x"], floor_idx["Pos_y"])
        ax[i, j].set_title(f"Floor {i * 4 + j + 1}", fontsize=14)

plt.show();
```

```python
pos_x_train_dict = uts_filter_train_data["Pos_x"]
pos_y_train_dict = uts_filter_train_data["Pos_y"]

pos_x_test_dict = uts_filter_test_data["Pos_x"]
pos_y_test_dict = uts_filter_test_data["Pos_y"]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

pos_x_train_dict.hist(bins=20, ax=ax[0, 0])
ax[0, 0].set_title("pos_x_train_dict", fontsize=15)

pos_y_train_dict.hist(bins=20, ax=ax[0, 1])
ax[0, 1].set_title("pos_y_train_dict", fontsize=15)


pos_x_test_dict.hist(bins=20, ax=ax[1, 0])
ax[1, 0].set_title("pos_x_test_dict", fontsize=15)

pos_y_test_dict.hist(bins=20, ax=ax[1, 1])
ax[1, 1].set_title("pos_y_test_dict", fontsize=15)

plt.show()
```
